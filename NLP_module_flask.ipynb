```python
# 라이브러리 임포트
import torch
from torch import nn
from flask import Flask, request, jsonify
from pyngrok import ngrok
from flask_cors import CORS
import json
import gluonnlp as nlp
import numpy as np
import pandas as pd
import tqdm
import time
import urllib.request
import tensorflow as tf
from kobert_tokenizer import KoBERTTokenizer
from transformers import BertModel
from transformers import AutoTokenizer
from transformers import TFGPT2LMHeadModel
```

```python
# CUDA GPU 사용
device = torch.device("cuda:0")

# KoBERT 모델 로드
bertmodel1 = BertModel.from_pretrained("skt/kobert-base-v1")
bertmodel2 = BertModel.from_pretrained("skt/kobert-base-v1")

# KoBERT 토크나이저 로드
tokenizer = KoBERTTokenizer.from_pretrained("skt/kobert-base-v1")

# KoBERT의 단어 사전 로드
vocab = nlp.vocab.BERTVocab.from_sentencepiece(tokenizer.vocab_file, padding_token='[PAD]')

# KoBERT 기반 1차 감정 분석 모델 정의
class BERTClassifier1(nn.Module):
    def __init__(self, bert, hidden_size=768, num_classes=3, dr_rate=None):
        super(BERTClassifier1, self).__init__()
        self.bert = bert
        self.dr_rate = dr_rate
        self.classifier = nn.Linear(hidden_size, num_classes)
        if dr_rate:
            self.dropout = nn.Dropout(p=dr_rate)  # 드롭아웃 추가

    def forward(self, token_ids, valid_length, segment_ids):
        attention_mask = self.gen_attention_mask(token_ids, valid_length)
        outputs = self.bert(
            input_ids=token_ids,
            token_type_ids=segment_ids.long(),
            attention_mask=attention_mask.float().to(token_ids.device),
        )

        # pooler가 `None`인 경우 대비
        pooler = outputs[1] if outputs[1] is not None else outputs[0][:, 0, :]

        if self.dr_rate:
            pooler = self.dropout(pooler)

        return self.classifier(pooler)

    def gen_attention_mask(self, token_ids, valid_length):
        attention_mask = torch.zeros_like(token_ids)
        for i, v in enumerate(valid_length):
            attention_mask[i][:v] = 1
        return attention_mask.float()

# KoBERT 기반 2차 감정 분석 모델 정의
class BERTClassifier2(nn.Module):
    def __init__(self, bert, hidden_size=768, num_classes=5, dr_rate=None):
        super(BERTClassifier2, self).__init__()
        self.bert = bert
        self.dr_rate = dr_rate
        self.classifier = nn.Linear(hidden_size, num_classes)
        if dr_rate:
            self.dropout = nn.Dropout(p=dr_rate)  # 드롭아웃 추가

    def forward(self, token_ids, valid_length, segment_ids):
        attention_mask = self.gen_attention_mask(token_ids, valid_length)
        outputs = self.bert(
            input_ids=token_ids,
            token_type_ids=segment_ids.long(),
            attention_mask=attention_mask.float().to(token_ids.device),
        )

        # pooler가 `None`인 경우 대비
        pooler = outputs[1] if outputs[1] is not None else outputs[0][:, 0, :]

        if self.dr_rate:
            pooler = self.dropout(pooler)

        return self.classifier(pooler)

    def gen_attention_mask(self, token_ids, valid_length):
        attention_mask = torch.zeros_like(token_ids)
        for i, v in enumerate(valid_length):
            attention_mask[i][:v] = 1
        return attention_mask.float()


    # 패딩된 토큰 무시하기 위한 Attention Mask 생성
    def gen_attention_mask(self, token_ids, valid_length):
        attention_mask = torch.zeros_like(token_ids)    # 토큰 ID와 동일한 크기의 텐서 생성
        for i, v in enumerate(valid_length):
            # 유효한 토큰 위치를 1로 설정
            attention_mask[i][:v] = 1
        return attention_mask.float()


# 저장된 모델 및 가중치 경로
MODEL_PATH_1 = "/content/drive/MyDrive/참빛설계학기/NLP_module_5/3_KoBERT_model/kobert_model.pth"
WEIGHTS_PATH_1 = "/content/drive/MyDrive/참빛설계학기/NLP_module_5/3_KoBERT_model/kobert_weights.pth"
BEST_MODEL_PATH = "/content/drive/MyDrive/참빛설계학기/NLP_module_5/3_KoBERT_model/best_model_old.pt"

MODEL_PATH_2 = "/content/drive/MyDrive/참빛설계학기/NLP_module_5/5_KoBERT_model/kobert_model.pth"
WEIGHTS_PATH_2 = "/content/drive/MyDrive/참빛설계학기/NLP_module_5/5_KoBERT_model/kobert_weights.pth"

# 1차 분류 모델
model_1 = BERTClassifier1(bertmodel1, dr_rate=0.5, num_classes=3).to(device)
# model_1.load_state_dict(torch.load(WEIGHTS_PATH_1, map_location=device))
model_1.load_state_dict(torch.load(BEST_MODEL_PATH, map_location=device))
model_1.eval()
print("1차 분류 KoBERT 모델 로드 완료")

# 2차 분류 모델
model_2 = BERTClassifier2(bertmodel2, dr_rate=0.5, num_classes=5).to(device)
model_2.load_state_dict(torch.load(WEIGHTS_PATH_2, map_location=device))
model_2.eval()
print("2차 분류 KoBERT 모델 로드 완료")


# 1차 분류용 (max_len=256)
def preprocess_text_1(text, tokenizer, max_len=256):
    inputs = tokenizer(
        text,
        max_length=max_len,
        padding="max_length",
        truncation=True,
        return_tensors="pt",
    )
    token_ids = inputs["input_ids"].to(device)
    segment_ids = inputs["token_type_ids"].to(device)
    valid_length = torch.tensor([len(tokenizer.tokenize(text))]).long().to(device)
    return token_ids, valid_length, segment_ids

# 2차 분류용 (max_len=64)
def preprocess_text_2(text, tokenizer, max_len=64):
    inputs = tokenizer(
        text,
        max_length=max_len,
        padding="max_length",
        truncation=True,
        return_tensors="pt",
    )
    token_ids = inputs["input_ids"].to(device)
    segment_ids = inputs["token_type_ids"].to(device)
    valid_length = torch.tensor([len(tokenizer.tokenize(text))]).long().to(device)
    return token_ids, valid_length, segment_ids

# KoGPT2 모델 로드
KOGPT2_PATH = "/content/drive/MyDrive/참빛설계학기/NLP_module_5/KoGPT2_model"
tokenizer_gpt2 = AutoTokenizer.from_pretrained(KOGPT2_PATH)
model_gpt2 = TFGPT2LMHeadModel.from_pretrained(KOGPT2_PATH)
```

```python
import re

def preprocess_for_kobert(text):
    # 감정 강조 부호 제거
    text = re.sub(r"[ㅋㅎㅠㅜㄷㅉㅊㅡㄱㄴㄷㄹㅁㅂㅅㅇㅈㅊㅌㅍㄲ]+", "", text)
    text = re.sub(r"[~♡♥^;*_/-⊙]+", "", text)
    text = re.sub(r"[.,!?\"]", "", text)
    text = re.sub(r"\.\.\.+", "", text)
    return text.strip()
```

```python
app = Flask(__name__)
CORS(app)
ngrok.set_auth_token("2wlbcgDh7SpDbt289vzmWBglpLb_6wEp1AziwiSmetL8pH2wQ")  # 본인의 토큰으로 설정

STOP_WORDS = {"그만하자", "그만 하자", "그만하자요", "그만하자고", "이제 됐어"}

def normalize(text):
    print(f"[DEBUG] 원본 텍스트: {text}")
    norm = re.sub(r"[^\w\s]", "", text.lower())
    print(f"[DEBUG] 정규화 텍스트: {norm}")
    return norm

def is_stop_intent(text):
    norm = normalize(text)
    for stop in STOP_WORDS:
        if stop.replace(" ", "") in norm.replace(" ", ""):
            return True
    return False

@app.route("/process/audio", methods=["POST"])
def full_pipeline():
    if 'audio' not in request.files:
        return jsonify({"error": "오디오 파일이 필요합니다."}), 400

    file = request.files['audio']
    history_text = request.form.get("history", "")  # 클라이언트에서 문자열 형태로 전달됨
    history_list = history_text.strip().split("\n") if history_text else []

    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp:
        file.save(tmp.name)
        print(f"오디오 저장: {tmp.name}")

        # 1. Whisper STT
        stt_result = whisper_model.transcribe(
            tmp.name,
            language="ko",
            fp16=torch.cuda.is_available(),
            no_speech_threshold=0.3,
            condition_on_previous_text=False
        )
        text = stt_result["text"].strip()
        print(f"Whisper 결과: {text}")

    # 2. 텍스트가 비었을 경우
    if not text:
        print("[DEBUG] Whisper 결과가 비어 있습니다.")
        return jsonify({
            "transcribed_text": "",
            "stop_intent": False,
            "emotion": "",
            "response": ""
        })

    # 3. 종료 의도 감지
    if is_stop_intent(text):
        print("[INFO] 종료 의도 감지됨.")
        return jsonify({
            "transcribed_text": text,
            "stop_intent": True
        })

    # 4. 감정 분석
    clean_text = preprocess_for_kobert(text)

    # 1차 감정 분석 (중립/부정/기쁨)
    token_ids_1, valid_length_1, segment_ids_1 = preprocess_text_1(clean_text, tokenizer)
    with torch.no_grad():
        logits_1 = model_1(token_ids_1, valid_length_1, segment_ids_1)
        pred_1 = torch.argmax(logits_1, dim=1).item()
    labels_1 = {0: "중립", 1: "부정", 2: "기쁨"}
    primary_emotion = labels_1[pred_1]

    # 2차 감정 분석 (부정일 경우)
    if primary_emotion == "부정":
        token_ids_2, valid_length_2, segment_ids_2 = preprocess_text_2(clean_text, tokenizer)
        with torch.no_grad():
            logits_2 = model_2(token_ids_2, valid_length_2, segment_ids_2)
            pred_2 = torch.argmax(logits_2, dim=1).item()
        label_dict = {0: "분노", 1: "슬픔", 2: "불안", 3: "상처", 4: "놀람"}
        emotion = label_dict[pred_2]
    else:
        emotion = primary_emotion

    # 5. KoGPT2 응답 생성
    history_trimmed = "\n".join(history_list[-3:])  # 최대 3줄만 유지
    conversation_input = history_trimmed + f"\n<usr>{text}<sys>"

    input_ids = [tokenizer_gpt2.bos_token_id] + tokenizer_gpt2.encode(conversation_input)
    input_ids = tf.convert_to_tensor([input_ids])
    output = model_gpt2.generate(input_ids, max_length=100, do_sample=True, top_k=20)
    sentence = tokenizer_gpt2.decode(output[0].numpy().tolist())
    response_text = sentence.split('<sys>')[-1].replace('</s>', '').strip()

    return jsonify({
        "transcribed_text": text,
        "stop_intent": False,
        "emotion": emotion,
        "response": response_text
    })

# 서버 실행 (ngrok 연결 포함)
public_url = ngrok.connect(5000)
print(f"[INFO] ngrok public URL: {public_url}")

app.run(host="0.0.0.0", port=5000)
```
